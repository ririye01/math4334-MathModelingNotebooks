# Demonstration: Semi-Log plots

First we have our standard header with common import statements, and a statement to make plots look nice.

```python
# standard import statements
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.optimize as opt

import classlib4334

# jupyter-specific display instructions
mpl.rcParams['font.size'] = 11
mpl.rcParams['lines.linewidth'] = 2
%config InlineBackend.figure_format = 'retina'
```

We are going to fit several things to lines here, so we'll define a line function once to re-use throughout the exercise.

```python
def line(x, a, b):
    return a + b*x
```

## Exercise 1

```python
excel_file = 'data/demo-semilog-1.xlsx'
mydata = pd.read_excel(excel_file)
x1 = np.array(mydata['x'])
y1 = np.array(mydata['y'])

plt.figure()
plt.plot(x1,y1, 'bs')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Original Data')
plt.tight_layout()
plt.show()
```

## Exercise 2

```python
excel_file = 'data/demo-semilog-2.xlsx'
mydata = pd.read_excel(excel_file)
x2 = np.array(mydata['x'])
y2 = np.array(mydata['y'])

plt.figure()
plt.plot(x2,y2, 'bs')
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Original Data')

plt.tight_layout()
plt.show()
```

```python
tx2 = np.log10(x2)
ty2 = np.log10(y2)

plt.figure()
plt.plot(np.log10(x2),np.log10(y2), 'bs')
plt.xlabel('log10(x)')
plt.ylabel('log10(y)')
plt.title(r'Log/Log Transform of Data')

plt.tight_layout()
plt.show()
```

```python
fit2, cov2 = classlib4334.fit_and_plot(line, tx2, ty2, xlabel="log10(x)", ylabel="log10(y)")
```

```python
plt.figure()
plt.plot(tx2, y2, 'bs')
plt.xlabel('log10(x)')
plt.ylabel('y')
plt.title(r'Semi-Log x Transform of Data')

plt.tight_layout()
plt.show()
```

```python
tx2 = np.log10(x2)

fit2a, cov2a = classlib4334.fit_and_plot(line, tx2, y2, xlabel="log10(x)")
```

$y = 3.0 + \log_{10}(x)$
